#Setting the x and y axis values

```
import numpy as np
import matplotlib.pyplot as plt


x = df['Date']
y = df['ClearSkyUVI']
```

# Graph of the month of January 2006, each point is a day

```

plt.scatter(x,y) # matplotlib.pylot scatterplot function
plt.xlim(20060101, 20060131) # Plots 1/1/2006 through 1/31/06 --> one month
plt.ylim(0,13) # Sets range for Clear Sky UV Index
plt.xlabel('Day')
plt.ylabel('Clear Sky UVI')
plt.show()
```

# Graph of Years 2006 to 2016, each vertical line (vertically stretched bell curve) is 1 year

```
plt.scatter(x,y)
plt.xlabel('Year')
plt.ylabel('Clear Sky UVI')
plt.xlim(20060101, 20161231) # Range for year of 2006
plt.ylim(0,12)
plt.show()
```

#Linear Regression Prediction over 2006

Predicts the UVI between Jan 1 2006 to Jan 31 2006

```
from sklearn.linear_model import LinearRegression
from sklearn.model_selection import cross_val_score

model = LinearRegression(fit_intercept = False, copy_X = True)
model.fit(x.values.reshape(3836,1),y) # variable.values.reshape(a, b) turns variable into a 2-D array

for x in range (20060101, 20060131):
  print model.predict(x)
  model.predict(x)
  
 #Current value returned for ANY date input(x) is relatively the same because linear regression poorly fits the data.
```
